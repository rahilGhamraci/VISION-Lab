```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math



# Lire l'image
image_path = '../img.jfif'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

th = 0

type_th = 0


def afficher():

    imgContour = np.zeros_like(img) # recomendations : use adequate type for the operation , nnormalization of values so the remain in the same interval

    grad_x = img[:,:img.shape[1]-1] - img[:,1:]

    grad_y = img[:img.shape[0]-1,:] - img[1:,:]

    grad_x = np.pad(grad_x, ((0, 0), (0, 1)), mode='constant') #handling last column
    grad_y = np.pad(grad_y, ((0, 1), (0, 0)), mode='constant') #handling last row


    grad = np.sqrt(grad_x**2 + grad_y**2)


    imgContour[grad > th] = 255

    imgContour[grad < th] = 0

  
    cv2.imshow('img',imgContour)

def change_th(x):
    global th
    th = x
    afficher()


cv2.namedWindow("img")
cv2.createTrackbar("thresh","img",0,256,change_th)

afficher()

#cv2.createTrackbar("type","img",0,4,change_type)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
